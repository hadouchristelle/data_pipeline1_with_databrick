#Read and writing dataset from bronze

##crm_cust


```python
df = spark.read.option("header", "true").option("inferSchema", "true").csv("/Volumes/workspace/bronze/vol_project/source_crm/cust_info.csv")
```

```python
df.write.mode("overwrite").saveAsTable("workspace.bronze.crm_cust")
```

##check delta table


```python
%sql
select*from workspace.bronze.crm_cust
```
