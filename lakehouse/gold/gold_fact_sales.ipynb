#Transformatiom

```python
query = """
SELECT
    sd.order_number,
    pr.product_number,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM silver.crm_sales sd
LEFT JOIN gold.dim_product pr
    ON sd.product_number = pr.product_key
LEFT JOIN gold.dim_customer cu
    ON sd.customer_id = cu.customer_id;
"""
df = spark.sql(query)
```

#writing gold table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
